```python
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import RidgeClassifier, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder, OrdinalEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import StackingClassifier, VotingClassifier, RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.feature_selection import RFECV, RFE, SelectFromModel
from imblearn.over_sampling import SMOTE
```

```python
data = pd.read_csv("ЛР 1 - Данные.csv")
data.head()
```

```python
data = data.drop(columns = ["Person ID"])
```

```python
data.info()
```

```python
data = data.fillna({"Sleep Disorder": "Нет"})
data.Gender = data.Gender.map({"Female": 1,
                "Male": 0})
data["Sleep Disorder"] = data["Sleep Disorder"].map({"Нет": 0,
                                                     "Insomnia": 1,
                                                     "Sleep Apnea": 2})
data["BMI Category"] = data["BMI Category"].replace({"Normal Weight": "Normal"})
data["Blood Pressure Sist"] = data["Blood Pressure"].apply(lambda x: int(x.split("/")[0]))
data["Blood Pressure Diast"] = data["Blood Pressure"].apply(lambda x: int(x.split("/")[1]))
data["Pulse Pressure"] = data["Blood Pressure Sist"] - data["Blood Pressure Diast"]
data = data.drop(columns = "Blood Pressure")
```

```python
data.info()
```

```python
X = data.drop(columns = ["Sleep Disorder"])
y = data["Sleep Disorder"]
```

```python
categories = [["Normal", "Overweight", "Obese"]]
oe_BMI = OrdinalEncoder(categories = categories)
X["BMI Category"] = oe_BMI.fit_transform(X[["BMI Category"]])
```

```python
ohe_job = OneHotEncoder(sparse_output = False, handle_unknown='ignore')
occupation = ohe_job.fit_transform(X[["Occupation"]])
columns_name = ohe_job.get_feature_names_out()
X = X.join(pd.DataFrame(occupation, columns = columns_name, index = X.index)).drop(columns = "Occupation")
```

```python
ss = StandardScaler()
ss.fit(X)
X = pd.DataFrame(ss.transform(X), columns = X.columns, index = X.index)
```

```python
cross_score = pd.DataFrame(np.zeros((3, 2)), 
                           index = ["Stacking", "Voting", "XGBoost"],
                           columns = ["Не перемешанные фолды", "Перемешанные фолды"])
```

```python
stacking_r = StackingClassifier([("logistic", LogisticRegression(max_iter = 1000, random_state = 100, class_weight='balanced')), 
                                 ("xgboost", XGBClassifier(random_state = 100)), ("forest", RandomForestClassifier(random_state = 100))])
voting_r = VotingClassifier([("logistic", LogisticRegression(max_iter = 1000, random_state = 100, class_weight='balanced')), 
                             ("xgboost", XGBClassifier(random_state = 100)), ("forest", RandomForestClassifier(random_state = 100))], voting = "soft")
xgb = XGBClassifier(random_state = 100)
```

```python
models = [stacking_r, voting_r, xgb]
```

```python
for i, model in enumerate(models):
    cross_score.iloc[i, 0] = cross_val_score(model, X, y, cv = 5, scoring = "f1_macro").mean()

cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 100)
for i, model in enumerate(models):
    cross_score.iloc[i, 1] = cross_val_score(model, X, y, cv = cv, scoring = "f1_macro").mean()
```

```python
cross_score
```

```python
X = data.drop(columns = ["Sleep Disorder"])
y = data["Sleep Disorder"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, stratify=y, random_state = 100, shuffle = True)
```

```python
categories = [["Normal", "Overweight", "Obese"]]
oe_BMI = OrdinalEncoder(categories = categories)
x_train["BMI Category"] = oe_BMI.fit_transform(x_train[["BMI Category"]])
x_test["BMI Category"] = oe_BMI.transform(x_test[["BMI Category"]])
ohe_job = OneHotEncoder(sparse_output = False, handle_unknown='ignore', 
                        categories = [list(x_train["Occupation"].value_counts().index[:7])])
occupation = ohe_job.fit_transform(x_train[["Occupation"]])
columns_name = ohe_job.get_feature_names_out()
x_train = x_train.join(pd.DataFrame(occupation, columns = columns_name, index = x_train.index)).drop(columns = "Occupation")
x_test = x_test.join(pd.DataFrame(ohe_job.transform(x_test[["Occupation"]]), columns = columns_name, index = x_test.index)).drop(columns = "Occupation")
ss = StandardScaler()
ss.fit(x_train)
x_train = pd.DataFrame(ss.transform(x_train), columns = x_train.columns, index = x_train.index)
x_test = pd.DataFrame(ss.transform(x_test), columns = x_test.columns, index = x_test.index)
```

```python
x_train_samled, y_train_samled = SMOTE(random_state = 100, k_neighbors = 3).fit_resample(x_train, y_train)
```

```python
stacking_r = StackingClassifier([("logistic", LogisticRegression(max_iter = 1000, random_state = 100, class_weight='balanced')), 
                                 ("xgboost", XGBClassifier(random_state = 100)), ("forest", RandomForestClassifier(random_state = 100))])
voting_r = VotingClassifier([("logistic", LogisticRegression(max_iter = 1000, random_state = 100, class_weight='balanced')), 
                                 ("xgboost", XGBClassifier(random_state = 100)), ("forest", RandomForestClassifier(random_state = 100))], 
                            voting = "soft")
xgb = XGBClassifier(random_state = 100)
models = [stacking_r, voting_r, xgb]
```

```python
model_score = pd.DataFrame(np.zeros((3, 6)), 
                           index = ["Stacking", "Voting", "XGBoost"],
                           columns = ["Train_Macro", "Test_Macro", "Train_Macro_best_paramets", "Test_Macro_best_paramets", 
                                      "Train_Macro_important_features", "Test_Macro_important_features"])
```

```python
for i, model in enumerate(models):
    model.fit(x_train_samled, y_train_samled)
    model_score.iloc[i, 0] = f1_score(y_train, model.predict(x_train), average = "macro")
    model_score.iloc[i, 1] = f1_score(y_test, model.predict(x_test), average = "macro")
```

```python
model_score
```

```python
cv_strat = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)
```

```python
par = param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 4, 5],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
    'gamma': [0, 0.1] 
}

gs_XGBoost = GridSearchCV(XGBClassifier(random_state = 100), par, cv = cv_strat, scoring = "f1_macro")
gs_XGBoost.fit(x_train_samled, y_train_samled)
```

```python
var = np.concat([np.logspace(-2, 2, 20), [1.0]])
par = {"C": var}

gs_logistic = GridSearchCV(LogisticRegression(max_iter = 1000, random_state = 100, class_weight='balanced'), par, cv = cv_strat, scoring = "f1_macro")
gs_logistic.fit(x_train_samled, y_train_samled)

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'ccp_alpha': [0, 0.001, 0.01, 0.1]
}
gs_random_forest = GridSearchCV(RandomForestClassifier(random_state = 100), param_grid, cv = cv_strat, scoring = "f1_macro")
gs_random_forest.fit(x_train_samled, y_train_samled)
```

```python
par = {'final_estimator': [gs_logistic.best_estimator_, gs_random_forest.best_estimator_, gs_XGBoost.best_estimator_],
       'stack_method': ['predict_proba', 'predict'],
       'passthrough': [False, True]
      }

gs_stack = GridSearchCV(StackingClassifier([("logistic", gs_logistic.best_estimator_), 
                                            ("xgboost", gs_XGBoost.best_estimator_), ("forest", gs_random_forest.best_estimator_)]), 
                        par, cv = cv_strat, scoring = "f1_macro")
gs_stack.fit(x_train_samled, y_train_samled)
```

```python
par = {
    'voting': ['hard', 'soft'],
    'weights': [[1, 1, 1], [2, 1, 1], [1, 2, 1], [1, 1, 2], [2, 2, 1], [2, 1, 2], [1, 2, 2]]
}

gs_vote = GridSearchCV(VotingClassifier([("logistic", gs_logistic.best_estimator_), 
                                         ("xgboost", gs_XGBoost.best_estimator_), 
                                         ("forest", gs_random_forest.best_estimator_)]), 
                       par, cv = cv_strat, scoring = "f1_macro")
gs_vote.fit(x_train_samled, y_train_samled)
```

```python
models = [gs_stack.best_estimator_, gs_vote.best_estimator_, gs_XGBoost.best_estimator_]
for i, model in enumerate(models):
    model_score.iloc[i, 2] = f1_score(y_train, model.predict(x_train), average = "macro")
    model_score.iloc[i, 3] = f1_score(y_test, model.predict(x_test), average = "macro")
```

```python
model_score
```

```python
selector = SelectFromModel(estimator = gs_XGBoost.best_estimator_, threshold = "median")
selector.fit(x_train_samled, y_train_samled)
print("SelectFromModel")
print(selector.threshold_)
select_columns = selector.get_feature_names_out()
print(select_columns)

print("RFE")
rfe = RFE(estimator = gs_XGBoost.best_estimator_, n_features_to_select = 10)
rfe.fit(x_train_samled, y_train_samled)
rfe_columns = rfe.get_feature_names_out()
print(rfe_columns)

rfecv = RFECV(estimator = gs_XGBoost.best_estimator_, 
              step = 1, cv = cv_strat, scoring = "f1_macro", min_features_to_select = 1)
rfecv.fit(x_train_samled, y_train_samled)
print("RFECV")
print(rfecv.n_features_)
rfecv_columns = rfecv.get_feature_names_out()
print(rfecv_columns)
```

```python
xg_boost_model = gs_XGBoost.best_estimator_
xg_boost_model.fit(x_train_samled[select_columns], y_train_samled)
print("SelectFromModel")
print(f1_score(y_train, xg_boost_model.predict(x_train[select_columns]), average = "macro"))
print(f1_score(y_test, xg_boost_model.predict(x_test[select_columns]), average = "macro"))

print("RFE")
xg_boost_model = gs_XGBoost.best_estimator_
xg_boost_model.fit(x_train_samled[rfe_columns], y_train_samled)
print(f1_score(y_train, xg_boost_model.predict(x_train[rfe_columns]), average = "macro"))
print(f1_score(y_test, xg_boost_model.predict(x_test[rfe_columns]), average = "macro"))
```

```python
selector = SelectFromModel(estimator = gs_logistic.best_estimator_, threshold = "median")
selector.fit(x_train_samled, y_train_samled)
print("SelectFromModel")
print(selector.threshold_)
select_columns = selector.get_feature_names_out()
print(select_columns)

print("RFE")
rfe = RFE(estimator = gs_logistic.best_estimator_, n_features_to_select = 10)
rfe.fit(x_train_samled, y_train_samled)
rfe_columns_xgboost = rfe.get_feature_names_out()
print(rfe_columns)

rfecv = RFECV(estimator = gs_logistic.best_estimator_, 
              step = 1, cv = cv_strat, scoring = "f1_macro", min_features_to_select = 1)
rfecv.fit(x_train_samled, y_train_samled)
print("RFECV")
print(rfecv.n_features_)
rfecv_columns = rfecv.get_feature_names_out()
print(rfecv_columns)
```

```python
logistic_model = gs_logistic.best_estimator_
logistic_model.fit(x_train_samled[select_columns], y_train_samled)
print("SelectFromModel")
print(f1_score(y_train, logistic_model.predict(x_train[select_columns]), average = "macro"))
print(f1_score(y_test, logistic_model.predict(x_test[select_columns]), average = "macro"))

print("RFE")
logistic_model = gs_XGBoost.best_estimator_
logistic_model.fit(x_train_samled[rfe_columns], y_train_samled)
print(f1_score(y_train, logistic_model.predict(x_train[rfe_columns]), average = "macro"))
print(f1_score(y_test, logistic_model.predict(x_test[rfe_columns]), average = "macro"))
```

```python
selector = SelectFromModel(estimator = gs_random_forest.best_estimator_, threshold = "median")
selector.fit(x_train_samled, y_train_samled)
print("SelectFromModel")
print(selector.threshold_)
select_columns = selector.get_feature_names_out()
print(select_columns)

print("RFE")
rfe = RFE(estimator = gs_random_forest.best_estimator_, n_features_to_select = 10)
rfe.fit(x_train_samled, y_train_samled)
rfe_columns = rfe.get_feature_names_out()
print(rfe_columns)

rfecv = RFECV(estimator = gs_random_forest.best_estimator_, 
              step = 1, cv = cv_strat, scoring = "f1_macro", min_features_to_select = 1)
rfecv.fit(x_train_samled, y_train_samled)
print("RFECV")
print(rfecv.n_features_)
rfecv_columns = rfecv.get_feature_names_out()
print(rfecv_columns)
```

```python
forest_model = gs_random_forest.best_estimator_
forest_model.fit(x_train_samled[select_columns], y_train_samled)
print("SelectFromModel")
print(f1_score(y_train, forest_model.predict(x_train[select_columns]), average = "macro"))
print(f1_score(y_test, forest_model.predict(x_test[select_columns]), average = "macro"))

print("RFE")
forest_model = gs_random_forest.best_estimator_
forest_model.fit(x_train_samled[rfe_columns], y_train_samled)
print(f1_score(y_train, forest_model.predict(x_train[rfe_columns]), average = "macro"))
print(f1_score(y_test, forest_model.predict(x_test[rfe_columns]), average = "macro"))

print("RFECV")
forest_model = gs_random_forest.best_estimator_
forest_model.fit(x_train_samled[rfecv_columns], y_train_samled)
print(f1_score(y_train, forest_model.predict(x_train[rfecv_columns]), average = "macro"))
print(f1_score(y_test, forest_model.predict(x_test[rfecv_columns]), average = "macro"))
```
