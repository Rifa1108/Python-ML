```python
data = data.drop(columns = "Date")
data = data.fillna(data.mode().iloc[0])
data.info()
```

```python
ohe = OneHotEncoder(sparse_output = False, drop = "first")
data = data.join(pd.DataFrame(ohe.fit_transform(data[["Ship_Type","Engine_Type"]]), columns = ohe.get_feature_names_out(), index = data.index)).drop(columns = ["Ship_Type", "Engine_Type"])
data.head()
```

```python
print(data["Route_Type"].value_counts())
print(data["Maintenance_Status"].value_counts())
print(data["Weather_Condition"].value_counts())
```

```python
data["Route_Type"] = data["Route_Type"].replace({"Coastal": "Short-haul",
                                                 "Transoceanic": "Long-haul"})                                             
```

```python
categories = [["Short-haul", "Long-haul"], ["Good", 'Fair', "Critical"], ["Calm", "Moderate", "Rough"]]
oe = OrdinalEncoder(categories = categories)
data[["Route_Type", "Maintenance_Status", "Weather_Condition"]] = oe.fit_transform(data[["Route_Type", "Maintenance_Status", "Weather_Condition"]])
data.info()
```

```python
ss = StandardScaler()
data = pd.DataFrame(ss.fit_transform(data), columns = data.columns, index = data.index)
data.head()
```

```python
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_) 

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.xlabel('Количество кластеров (k)')
plt.ylabel('Инерция (WCSS)')
plt.title('Метод локтя для оптимального k')
plt.xticks(k_range)
plt.grid(True)

print(np.diff(inertia))

plt.axvline(x=5, color='red', linestyle=':', label='Предполагаемый оптимум')
plt.legend()
plt.show()
```

```python
silhouette_scores = []
k_range = range(2, 20)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(data)
    score = silhouette_score(data, cluster_labels)
    silhouette_scores.append(score)
    print(f"k = {k}, Silhouette Score = {score:.4f}")

plt.figure(figsize=(8, 5))
plt.plot(k_range, silhouette_scores, marker='o', linestyle='--')
plt.xlabel('Количество кластеров (k)')
plt.ylabel('Силуэтный коэффициент')
plt.title('Выбор оптимального k по силуэту')
plt.xticks(k_range)
plt.grid(True)

plt.axvline(x=3, color='red', linestyle=':', label='Предполагаемый оптимум')
plt.legend()
plt.show()
```

```python
def compute_inertia(X, labels, centers):
    inertia = 0
    for i, center in enumerate(centers):
        cluster_points = X[labels == i]
        inertia += np.sum((cluster_points - center) ** 2)
    return inertia

def gap_statistic(X, kmax=10, n_refs=10):
    gaps = []
    s_k = []
    X_min, X_max = np.min(X, axis=0), np.max(X, axis=0)
    
    for k in range(1, kmax + 1):
        print(f"Обработка k={k}...")
        
        # Кластеризация реальных данных
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(X)
        labels = kmeans.labels_
        centers = kmeans.cluster_centers_
        log_wcss_real = np.log(compute_inertia(X, labels, centers))
        
        # Кластеризация случайных данных
        log_wcss_refs = []
        for _ in range(n_refs):
            X_ref = np.random.uniform(low=X_min, high=X_max, size=X.shape)
            kmeans_ref = KMeans(n_clusters=k, random_state=None, n_init=10)
            kmeans_ref.fit(X_ref)
            labels_ref = kmeans_ref.labels_
            centers_ref = kmeans_ref.cluster_centers_
            log_wcss_ref = np.log(compute_inertia(X_ref, labels_ref, centers_ref))
            log_wcss_refs.append(log_wcss_ref)
        
        mean_log_wcss_ref = np.mean(log_wcss_refs)
        std_log_wcss_ref = np.std(log_wcss_refs)
        gap = mean_log_wcss_ref - log_wcss_real
        gaps.append(gap)
        s_k.append(std_log_wcss_ref * np.sqrt(1 + 1.0 / n_refs))
        
        print(f"  gap = {gap:.4f}")
    
    return gaps, s_k
```

```python
gaps, s_k = gap_statistic(data.values)

plt.figure(figsize=(10, 5))
plt.errorbar(range(1, len(gaps)+1), gaps, yerr=s_k, fmt='o-', capsize=5)
plt.xlabel('Количество кластеров')
plt.ylabel('Gap')
plt.title('Gap-статистика')
plt.grid(True)
plt.show()

# Оптимальное число кластеров
optimal_k = np.argmax(gaps) + 1
print(f"Оптимальное число кластеров: {optimal_k}")
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

Z = linkage(data, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='level', p=5)
plt.title('Дендрограмма (метод Уорда)')
plt.xlabel('Индексы объектов (или размеры кластеров)')
plt.ylabel('Расстояние')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 42, n_init = 10)
y_kmeans = kmeans.fit_predict(data)
y_agglomerative = fcluster(Z, t=3, criterion='maxclust')
db = DBSCAN(eps = 3.5, min_samples=5)
y_dbscan = db.fit_predict(data)
set(y_dbscan)
```

```python
for m in range(5, 30, 4):
    for e in np.arange(2, 4, 0.2): 
        db = DBSCAN(eps = e, min_samples=m)
        y_dbscan = db.fit_predict(data)
        n_clusters = len(set(y_dbscan))
        
        if n_clusters < 2:
            continue
        print(m, e)
        print(silhouette_score(data, y_dbscan))
```

```python
from sklearn.metrics import (silhouette_score, calinski_harabasz_score, davies_bouldin_score)

inertia = kmeans.inertia_
silhouette_avg = silhouette_score(data, y_kmeans)
ch_score = calinski_harabasz_score(data, y_kmeans)
db_score = davies_bouldin_score(data, y_kmeans)
print(f"Inertia (WCSS): {inertia:.2f}")
print(f"Silhouette Score: {silhouette_avg:.4f}")
print(f"Calinski-Harabasz Index: {ch_score:.2f}")
print(f"Davies-Bouldin Index: {db_score:.4f}")
```

```python
silhouette_avg = silhouette_score(data, y_agglomerative)
ch_score = calinski_harabasz_score(data, y_agglomerative)
db_score = davies_bouldin_score(data, y_agglomerative)

print(f"Silhouette Score: {silhouette_avg:.4f}")
print(f"Calinski-Harabasz Index: {ch_score:.2f}")
print(f"Davies-Bouldin Index: {db_score:.4f}")
```

```python
silhouette_avg = silhouette_score(data, y_dbscan)
ch_score = calinski_harabasz_score(data, y_dbscan)
db_score = davies_bouldin_score(data, y_dbscan)

print(f"Silhouette Score: {silhouette_avg:.4f}")
print(f"Calinski-Harabasz Index: {ch_score:.2f}")
print(f"Davies-Bouldin Index: {db_score:.4f}")
```
